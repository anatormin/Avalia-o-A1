```python

print("Contador com for:")
for numero in range(1, 101):
    if numero % 2 == 0:
        print(numero)

print("\nContador com while:")

numero = 1
while numero <= 100:
    if numero % 2 == 0:
        print(numero)
    numero += 1
```
